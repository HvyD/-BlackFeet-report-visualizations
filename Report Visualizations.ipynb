```python
from os import path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import os

from wordcloud import WordCloud, STOPWORDS

# get data directory (using getcwd() is needed to support running example in generated IPython notebook)
d = path.dirname(__file__) if "__file__" in locals() else os.getcwd()

# read the mask image
text = open(path.join(d, 'data/text.txt')).read()
#mask = np.array(Image.open(path.join(d, "images/buff.jpeg")))
mask = np.array(Image.open(path.join(d, "images/buff.jpg")))
#mask = np.array(Image.open(path.join(d, "images/default.jpg")))
#stopwords = set(STOPWORDS)
#stopwords.add("said")

wc = WordCloud(background_color="white", max_words=2000, mask=mask,
                contour_width=3, contour_color='steelblue')

# generate word cloud
wc.generate(text)

# store to file
wc.to_file(path.join(d, "images/WcBuffWc.png"))

# show
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.figure()
plt.imshow(mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
text = open(path.join(d, 'data/text.txt')).read()

d = getcwd()

mask = np.array(Image.open(path.join(d, "images/buff.jpg")))
wc = WordCloud(background_color="white", max_words=500, mask=mask,max_font_size=90)
wc.generate(text)
image_colors = ImageColorGenerator(mask)
plt.figure(figsize=[10,8],dpi=400)
plt.tight_layout(pad=0)
plt.imshow(wc.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
import urllib
import requests
import numpy as np
import matplotlib.pyplot as plt

words = text
mask = np.array(Image.open(path.join(d, "images/buff.jpg")))
WordCloud(background_color="white", max_words=500, mask=mask,max_font_size=90)

# This function takes in your text and your mask and generates a wordcloud. 
def generate_wordcloud(words, mask):
    word_cloud = WordCloud(background_color="white", max_words=500, mask=mask,max_font_size=90).generate(words)

    plt.figure(figsize=(10,8))
    plt.imshow(word_cloud)
    plt.axis('off')
    plt.tight_layout(pad=0)
    plt.show()
    
#Run the following to generate your wordcloud
generate_wordcloud(words, mask)
```
